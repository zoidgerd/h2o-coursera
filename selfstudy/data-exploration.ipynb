```python
import h2o
import pandas as pd
import numpy as np
import h2o.grid

from h2o.estimators.glm import H2OGeneralizedLinearEstimator

pd.options.display.max_columns = 40
pd.options.display.max_rows = 40
```

```python
h2o.init()
```

```python
data = h2o.import_file('http://h2o-public-test-data.s3.amazonaws.com/smalldata/airlines/allyears2k_headers.zip')
dataframe = data.as_data_frame()
```

```python
data.summary()
```

```python
allColumns = data.columns
allColumns.remove("ArrDelay")

specificColumns = data.columns
specificColumns.remove("UniqueCarrier")
specificColumns.remove("TailNum")
specificColumns.remove("TaxiIn")
specificColumns.remove("TaxiOut")
specificColumns.remove("ArrDelay")

targetColumn = "IsArrDelayed"
```

```python
glm_model_exploration = H2OGeneralizedLinearEstimator(family= "binomial",
                                          model_id = "flight-pred-exploration")

glm_model_exploration.train(specificColumns, targetColumn, training_frame = data)
```

```python
glm_model_exploration
```

```python
train, test, validation = data.split_frame([0.8, 0.1])
```

```python
glm_model_real = H2OGeneralizedLinearEstimator(family="binomial", model_id="flight-pred-real")
glm_model_real.train(specificColumns, targetColumn, train, validation_frame=validation)
```

```python
glm_model_real.model_performance(valid=True)
```

```python
preds = glm_model_real.predict(test)
```
