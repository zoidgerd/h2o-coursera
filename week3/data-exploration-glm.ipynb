```python
import pandas as pd
import numpy as np
import h2o
from h2o.estimators.glm import H2OGeneralizedLinearEstimator
import csv

h2o.init()
```

```python
data = h2o.import_file('data/smoking.csv')
```

```python
data.head()
```

```python
# columns i want to inspect
x = ["C1", "C2"]

#column i want to predict
y = "C5"
```

```python
glm_model = H2OGeneralizedLinearEstimator(family= "poisson",
                                          model_id = "test")

glm_model.train(x, y, training_frame=data)
```

```python
glm_model
```

```python
glm_model2 = H2OGeneralizedLinearEstimator(family= "poisson",
                                          model_id = "test2")

glm_model2.train(["C2"], y, training_frame=data)
```
