```python
import h2o
import h2o.grid
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h2o.init()
```

```python
data = h2o.import_file('http://coursera.h2o.ai/cacao.882.csv')
```

```python
data.describe()
```

```python
train, valid, test = data.split_frame([0.7, 0.15], seed=420)
```

```python
X = data.columns
y = 'Maker Location'
X.remove(y)
```

```python
model = H2ODeepLearningEstimator(model_id="model1")
%time model.train(X, y, train, validation_frame=valid)
```

```python
model.model_performance(test)
```

```python
model2 = H2ODeepLearningEstimator(model_id="model2", epochs=70, hidden = [100, 100], stopping_metric='logloss', 
                                                  stopping_tolerance=0.001 ,stopping_rounds=4)
%time model2.train(X, y, train, validation_frame=valid)
```

```python
model2.model_performance(test)
```

```python
model1 = h2o.get_model("model1")
model2 = h2o.get_model("model2")
h2o.save_model(model1, './')
h2o.save_model(model2, './')
```
